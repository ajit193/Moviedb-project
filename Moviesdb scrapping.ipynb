```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.themoviedb.org/movie'  

header = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/99.0.4844.51 Safari/537.36'}

source_data = requests.get(url, headers = header).text
soup_data = BeautifulSoup(source_data, 'lxml')
```

```python
url_lst = []
for i in range(1, 51):
    page = f"?page={i}"
    url_lst.append(url + page)
url_lst
```

```python
all_movie_data = []

for i in url_lst:
    source_data_all = requests.get(i, headers = header).text
    soup_data_all = BeautifulSoup(source_data_all,'lxml')
    all_card = soup_data_all.find_all('div', class_='card style_1')
    
    for item in all_card:
        link = str(item.find('a'))[29:37]
        new_url = url + link
        movie = requests.get(new_url, headers = header).text
        movie_soup = BeautifulSoup(movie ,'lxml')
        
        #finding name ,release date,movie rating
        movie_header = movie_soup.find('section',id='original_header')
        
        
        #movie name
        try:
            movie_name = movie_header.find('h2').find('a').text
        except:
            movie_name = None
            
         #release date
        try:
            release_date = movie_header.find('span',class_='release').text.replace('\n','').strip().split(" ")[0]
        except:
            release_date = None
            
         #movie rating
        try:
            rating = movie_header.find('div',class_='user_score_chart')["data-percent"]
        except:
            rating = None
            
            #director , SOME Movies have many directors using loop
            
        try:
            production = movie_header.find('ol',class_='people no_image')
            profile = production.find_all('li',class_='profile')
            director = ""
            
            for i in range(0, len(profile)):
                character = str(profile[i].find_all('p')[1].text)
                if 'Director' in character:
                    director+=f"{profile[i].find('a').text},"
                    director = director.rstrip(',')
        except:
            director : None
                
                
       # dur has none value in pages where it doesnt exist hence has no attribute text
        try:
            duration = movie_header.find('span', class_='runtime').text.replace('\n','').strip()
        except:
            duration = None
            
            
                
         #genre movie
        try:
            genre = movie_header.find('span',class_='genres').text.replace('\n','').replace('\xa0','')
        
        except:
            genre = None
            
            
            
        #stored a dictionary   
        all_movie_disc = {
            'Movie name' : movie_name,
            'Relese data' : release_date,
            'Rating': rating,
            'Director' : director,
            'Duration' : duration,
            'Genre' : genre
         }
        
        all_movie_data.append(all_movie_disc)
all_movie_data
```

```python
df = pd.DataFrame(all_movie_data)
```
